```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def f(a,x,b):
    return a*x+b
```

```python
# Etapa I - modelar a função ax+b
x_values=np.linspace(-10,10,25)
y_values=f(3,x_values,1)

# Etapa II - aplicar o ruído gaussiano
#noise = np.random.normal(-2,2,25)
#y_noise=y_values+noise
std = 0.5*np.abs(y_values)
noise = std*np.random.rand(len(y_values))
y_noise = y_values + noise

# Etapa III - estimar "a" e "b" pelo Mínimo Quadrado
ym=np.mean(y_noise)
xm=np.mean(x_values)

S1=np.sum(x_values*(y_noise-ym))
S2=np.sum(x_values*(x_values-xm))

a1=S1/S2
b1=ym-a1*xm

y_ajus=a1*x_values+b1
R2 = np.sum((y_ajus-ym)**2)/np.sum((y_noise-ym)**2) #Calcular o coeficiente de determinação(bônus)
```

```python
font1 = {'family':'times new roman','color':'blue','size':20}
font2 = {'family':'times new roman','color':'blue','size':15}

plt.figure(figsize=(,7))
sns.set_style('darkgrid')
plt.rc('axes', labelsize=13)
plt.rc('xtick', labelsize=13)
plt.rc('ytick', labelsize=13)
plt.rc('legend', fontsize=12)
plt.plot(x_values,y_values,linewidth=1.1,linestyle='-',label=f'{a}x+{b}',color='#3d30a2') #plot f(x)
plt.scatter(x_values,y_noise,label='Ruído',color='#f875aa',linewidth=1.5,marker='.') #plot noise
plt.plot(x_values,y_ajus,'--',color='#4f6f52',label=f'{a1}x{b1}\n R²={np.round(R2,2)}') #plot regressão linear
plt.xlabel('Valores de X',**font2)
plt.ylabel('Valores de Y',**font2)
plt.title('Gráficos de f(x) e Regressão Linear',**font1)
plt.legend()
plt.show()
```
